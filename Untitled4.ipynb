```python
from sklearn.cluster import KMeans
import pandas as pd

# The same 6 students from the board
df = pd.DataFrame({
    'student' : ['A','B','C','D','E','F'],
    'math'    : [80, 75, 85, 30, 25, 35],
    'science' : [85, 80, 90, 25, 30, 20]
})

X = df[['math','science']]

# K=2 — find 2 groups
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans.fit(X)

df['group'] = kmeans.labels_
print(df)

print(f"\nCenter 1: {kmeans.cluster_centers_[0]}")
print(f"Center 2: {kmeans.cluster_centers_[1]}")
```

```python
from sklearn.cluster import KMeans
import pandas as pd
import mlflow
from sklearn.metrics import silhouette_score
```

```python
df = pd.DataFrame({
    'student' : ['A','B','C','D','E','F'],
    'math'    : [80, 75, 85, 30, 25, 35],
    'science' : [85, 80, 90, 25, 30, 20]
})

X = df[['math','science']]
print(df)
```

```python
mlflow.set_experiment("Student_Clustering")
```

```python
with mlflow.start_run(run_name="kmeans_k2"):

    # Train
    kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
    kmeans.fit(X)

    df['group'] = kmeans.labels_

    # Metrics
    inertia = kmeans.inertia_
    score   = silhouette_score(X, kmeans.labels_)

    # Log
    mlflow.log_param("k",            2)
    mlflow.log_param("n_students",   len(df))

    mlflow.log_metric("inertia",     round(inertia, 2))
    mlflow.log_metric("silhouette",  round(score, 3))

    print(df)
    print(f"\nCenter 1 : {kmeans.cluster_centers_[0]}")
    print(f"Center 2 : {kmeans.cluster_centers_[1]}")
    print(f"\nInertia  : {inertia:.2f}")
    print(f"Silhouette: {score:.3f}")
    print(f"\n Logged to MLflow")
```

```python
with mlflow.start_run(run_name="kmeans_k3"):

    kmeans3 = KMeans(n_clusters=3, random_state=42, n_init=10)
    kmeans3.fit(X)

    inertia = kmeans3.inertia_
    score   = silhouette_score(X, kmeans3.labels_)

    mlflow.log_param("k",           3)
    mlflow.log_param("n_students",  len(df))
    mlflow.log_metric("inertia",    round(inertia, 2))
    mlflow.log_metric("silhouette", round(score, 3))

    print(f"K=3 → Inertia: {inertia:.2f}  Silhouette: {score:.3f}")
    print(f"\n Logged to MLflow")
```

```python
runs = mlflow.search_runs(
    experiment_names=["Student_Clustering"]
)

print(runs[['tags.mlflow.runName',
            'params.k',
            'metrics.inertia',
            'metrics.silhouette']].to_string(index=False))
```

```python
# Step 1 — Fetch the page (same as your browser does)
url = "http://quotes.toscrape.com"
response = requests.get(url)

print(f"Status Code : {response.status_code}")
print(f"200 means   : Server said OK, here's your page")
print(f"\nFirst 500 characters of raw HTML:")
print(response.text[:500])
```

```python
# Step 2 — Parse the HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Find the page title
title = soup.find('title')
print(f"Page Title: {title.text}")

# Find ALL quotes on this page
quotes = soup.find_all('span', class_='text')
print(f"\nFound {len(quotes)} quotes on this page\n")

# Print first 3
for i, quote in enumerate(quotes[:3], 1):
    print(f"Quote {i}: {quote.text}")
```

```python
# Get quotes AND their authors together
quotes_data = []

quote_blocks = soup.find_all('div', class_='quote')

for block in quote_blocks:
    quote  = block.find('span', class_='text').text
    author = block.find('small', class_='author').text
    tags   = [t.text for t in block.find_all('a', class_='tag')]

    quotes_data.append({
        'quote'  : quote,
        'author' : author,
        'tags'   : ', '.join(tags)
    })

df_quotes = pd.DataFrame(quotes_data)
print(f"Scraped {len(df_quotes)} quotes with authors and tags\n")
print(df_quotes.to_string(index=False))
```

```python
# This website has 10 pages — let's scrape ALL of them
all_quotes = []

for page_num in range(1, 11):
    url = f"http://quotes.toscrape.com/page/{page_num}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    blocks = soup.find_all('div', class_='quote')

    for block in blocks:
        quote  = block.find('span', class_='text').text
        author = block.find('small', class_='author').text
        tags   = [t.text for t in block.find_all('a', class_='tag')]

        all_quotes.append({
            'quote'  : quote,
            'author' : author,
            'tags'   : ', '.join(tags),
            'page'   : page_num
        })

    print(f"✅ Page {page_num} scraped — {len(blocks)} quotes")
    time.sleep(0.5)   # be polite — don't hammer the server

df_all = pd.DataFrame(all_quotes)
print(f"\nTotal quotes collected: {len(df_all)}")
df_all.head()
```

```python
import matplotlib.pyplot as plt

# Who has the most quotes?
top_authors = df_all['author'].value_counts().head(10)

fig, ax = plt.subplots(figsize=(10, 5))
top_authors.plot(kind='bar', color='steelblue',
                 edgecolor='black', ax=ax)
ax.set_title('Most Quoted Authors', fontsize=14, fontweight='bold')
ax.set_xlabel('Author')
ax.set_ylabel('Number of Quotes')
ax.set_xticklabels(top_authors.index, rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Most popular tags
all_tags = df_all['tags'].str.split(', ').explode()
all_tags = all_tags[all_tags != '']

top_tags = all_tags.value_counts().head(10)

fig, ax = plt.subplots(figsize=(10, 5))
top_tags.plot(kind='barh', color='coral',
              edgecolor='black', ax=ax)
ax.set_title('Most Common Quote Tags', fontsize=14, fontweight='bold')
ax.set_xlabel('Count')
plt.tight_layout()
plt.show()
```
